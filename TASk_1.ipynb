```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing modified dataset (Modified in Excel)
indrice = pd.read_csv('Mod_data.csv')
```

```python
indrice
```

```python
indrice = pd.get_dummies(indrice, columns = ['State'])  #One Hot Encoding
indrice
```

```python
# Printing the list of columns
for col in indrice.columns:
  print(col)
```

```python
# Converting boolean values to integer values
for col in indrice.columns:
  if col != "Year" or col!= "Production" or col!="NextVal":
    indrice[col] = indrice[col].astype(int)

indrice
```

```python
#Creating a list of features which we are using for prediction
predictors = ["Year", "Production","State_ALL INDIA",
"State_Andhra Pradesh",
"State_Arunachal Pradesh",
"State_Assam",
"State_Bihar",
"State_Chhattisgarh",
"State_Goa",
"State_Gujarat",
"State_Haryana",
"State_Himachal Pradesh",
"State_Jammu & Kashmir",
"State_Jharkhand",
"State_Karnataka",
"State_Kerala",
"State_Madhya Pradesh",
"State_Maharashtra",
"State_Manipur",
"State_Meghalaya",
"State_Mizoram",
"State_NCT of Delhi",
"State_Nagaland",
"State_Odisha",
"State_Puducherry",
"State_Punjab",
"State_Rajasthan",
"State_Sikkim",
"State_Tamil Nadu",
"State_Telangana",
"State_Tripura",
"State_Uttar Pradesh",
"State_Uttarakhand",
"State_West Bengal"]

# List of features which we will be training our data against
target = ["NextVal"]
```

```python
# Training Data : Pre-2018
# Testing Data  : Post 2018 (Including)
train = indrice[indrice["Year"]<2018].copy()
test = indrice[indrice["Year"]>=2018].copy()
```

```python
train
```

```python
# Dividing the training and testing data into Training Features and Testing Features according to the split
X_train = train[predictors]
X_test = test[predictors]
y_train = train[target]
y_test = test[target]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Training a Linear Regression Model
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Creating a list of predictions by input of the testing features
predictions = reg.predict(X_test)
```

```python
len(predictions)
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(predictions, y_test)

error
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
score = np.sqrt(mean_squared_error(y_test, predictions))
print("The Root Mean Squared Error of our Model is {}".format(round(score, 2)))
```
