```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
indrice = pd.read_csv('Mod_data.csv')
```

```python
indrice
```

```python
indrice = pd.get_dummies(indrice, columns = ['State'])  #One Hot Encoding
indrice
```

```python
for col in indrice.columns:
  print(col)
```

```python
for col in indrice.columns:
  if col != "Year" or col!= "Production" or col!="NextVal":
    indrice[col] = indrice[col].astype(int)

indrice
```

```python
predictors = ["Year", "Production","State_ALL INDIA",
"State_Andhra Pradesh",
"State_Arunachal Pradesh",
"State_Assam",
"State_Bihar",
"State_Chhattisgarh",
"State_Goa",
"State_Gujarat",
"State_Haryana",
"State_Himachal Pradesh",
"State_Jammu & Kashmir",
"State_Jharkhand",
"State_Karnataka",
"State_Kerala",
"State_Madhya Pradesh",
"State_Maharashtra",
"State_Manipur",
"State_Meghalaya",
"State_Mizoram",
"State_NCT of Delhi",
"State_Nagaland",
"State_Odisha",
"State_Puducherry",
"State_Punjab",
"State_Rajasthan",
"State_Sikkim",
"State_Tamil Nadu",
"State_Telangana",
"State_Tripura",
"State_Uttar Pradesh",
"State_Uttarakhand",
"State_West Bengal"]

target = ["NextVal"]
```

```python
train = indrice[indrice["Year"]<2018].copy()
test = indrice[indrice["Year"]>=2018].copy()
```

```python
train
```

```python
X_train = train[predictors]
X_test = test[predictors]
y_train = train[target]
y_test = test[target]
```

```python
from sklearn.preprocessing import PolynomialFeatures

polynom = PolynomialFeatures(degree = 2)
X_polynom = polynom.fit_transform(X_train[predictors])

X_polynom
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_polynom, y_train[target])
```

```python
X_polynom_test = polynom.fit_transform(X_test[predictors])

X_polynom_test
```

```python
predictions_pol = reg.predict(X_polynom_test)
```

```python
len(predictions_pol)
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(predictions_pol, y_test)

error
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, predictions_pol)
print("The accuracy of our model is {}%".format(round(score, 2) *100))
```

```python
from sklearn.metrics import mean_absolute_error
score = mean_absolute_error(y_test, predictions_pol)
print("The Mean Absolute Error of our Model is {}".format(round(score, 2)))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
score = np.sqrt(mean_squared_error(y_test, predictions_pol))
print("The Root Mean Squared Error of our Model is {}".format(round(score, 2)))
```
